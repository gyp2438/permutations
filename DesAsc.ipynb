```python
def ascent(sigma):
    return [i+1 for i in range(len(sigma) - 1) if sigma[i] < sigma[i + 1]]
```

```python
n=2
even_cycles = [sigma for sigma in Permutations(2*n) if all(cycle % 2 == 0 for cycle in sigma.cycle_type()) == True]
odd_cycles = [sigma for sigma in Permutations(2*n) if all(cycle % 2 == 1 for cycle in sigma.cycle_type()) == True]

len(even_cycles) == len(odd_cycles)
```

```python
Des_even = {}
for sigma in even_cycles:
    des = sigma.descents()
    Des_even[sigma] = des

Asc_odd = {}
for sigma in odd_cycles:
    asc = [i+1 for i in range(len(sigma) - 1) if sigma[i] < sigma[i + 1]]
    Asc_odd[sigma] = asc
```

```python
#all possible des/asc sets. Supposedly they are equal to one another, so just run it in descent
# Convert each list to a tuple and create a set
unique_sets = {tuple(lst) for lst in Des_even.values()}
unique_lists = [list(t) for t in unique_sets]
```

```python
# set_label = {}
perm_by_sets = {}
perm_by_RSK = {}
for i in range(len(unique_lists)):
#for value in unique_lists:
    # set_label[i] = value
    even=[]
    even_cycle = []
    even_RSK = []
    odd=[]
    odd_cycle = []
    odd_RSK = []
    for k, v in Des_even.items():
        if v == unique_lists[i]:
            even.append(k)
            even_cycle.append(k.to_cycles())
            even_RSK.append(RSK(k)[1])
    for k, v in Asc_odd.items():
        if v == unique_lists[i]:
            odd.append(k)
            odd_cycle.append(k.to_cycles())
            odd_RSK.append(RSK(k)[1])
    perm_by_sets[i] = [even,even_cycle,odd,odd_cycle]
    perm_by_RSK[i] = [even_RSK, odd_RSK]
```

```python
Dfs = []
for i in range(len(unique_lists)):
    df = pd.DataFrame(perm_by_sets[i],index = ['even', 'even cycle', 'odd','odd cycle']).T
    df.index.name = str(unique_lists[i])
    Dfs.append(df)
```
